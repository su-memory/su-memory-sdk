```python
!pip install su-memory -q
import sys, traceback
try:
    from su_memory import SuMemory
    print('✅ su-memory 已加载')
    client = SuMemory()
    print('✅ SuMemory() 初始化成功')
except Exception:
    print('安装失败，查看错误：')
    traceback.print_exc()
```

## 添加记忆 + 因果链

```python
# 添加5条记忆
memories = [
    '项目ROI增长25%，预计收益翻倍',
    '市场波动加大，需优化风险控制',
    '团队扩大至50人，组织架构调整',
    '技术债务积累，架构重构',
    '客户NPS增长15点'
]
for m in memories:
    mid = client.add(m)
    print(f'添加: {mid} {m[:15]}...')

# 建立因果关联
for i in range(len(memories)-1):
    client.link(f'mem_{i+1}', f'mem_{i+2}')
print('\n因果链已建立')
```

## 语义检索

```python
# 查询记忆
results = client.query('投资回报', top_k=3)
print(f'\n找到 {len(results)} 条相关记忆\n')
for r in results:
    print(f'  [{r.memory_id}] {r.content[:20]}... (score={r.score:.2f})')
    print(f'    八卦={r.encoding.bagua} 五行={r.encoding.wuxing} 能量={r.encoding.energy}')
    print()
```

## 记忆统计

```python
stats = client.get_stats()
print(f"""总记忆数: {stats['total_memories']}""")
print(f"八卦分布: {stats['bagua_distribution']}""")
```
